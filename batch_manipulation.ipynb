Automatically pull all Git repositories.

```python
for path in Path('.').iterdir():
    if path.is_dir() and path.joinpath('.git').exists():
        cmd = f'git -C {path.resolve()} pull origin master'
        print('\n\n' + cmd)
        sp.run(cmd, shell=True, check=True)
```

Automatically commit and push all Git repositories.

```python
for path in Path('.').iterdir():
    if path.is_dir() and path.joinpath('.git').exists():
        path = path.resolve()
        cmd = f'''git -C {path} add *.md && \
                git -C {path} commit -m "update readme.md"
                git -C {path} push origin master
            '''
        # print('\n\n' + cmd)
        proc = sp.run(cmd, shell=True, capture_output=True)
        print(proc.stdout.decode())
        print(proc.stderr.decode())
```

Automatically change legacy Bash-style `` `some_command` `` to `$(some_command)`.

```python
def replace_shell_command(path_text: Union[Path, str]):
    if isinstance(path_text, Path):
        with path_text.open() as fin:
            text = fin.read()
        text = replace_shell_command(text)
        with path_text.open('w') as fout:
            fout.write(text)
    mapping = {
        '`nproc`': '$(nproc)',
        '`id -u`': '$(id -u)',
        '`id -un`': '$(id -un)',
        '`id -g`': '$(id -g)',
        '`pwd`': '$(pwd)',
        '`dirname $HOME`': '$(dirname $HOME)',
    }
    for key, val in mapping.items():
        path_text = path_text.replace(key, val)
    return path_text


for path in Path('.').iterdir():
    try:
        path = next(path.glob('*.md'))
        replace_shell_command(path)
    except:
        pass
```
